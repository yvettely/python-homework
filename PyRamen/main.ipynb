```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import pandas as pd
from pathlib import Path
import csv


# @TODO: Set file paths for menu_data.csv
menu_filepath = Path('../PyRamen/menu_data.csv')


# @TODO: Initialize list objects to hold our menu data
menu_list = []


# @TODO: Read in the menu data into the menu list
with open(menu_filepath, 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter = ',')
    next(csv_reader, None)
    
# @TODO: Initialize dict object to hold our key-value pairs of items and metrics

    item_list = []
    category_list = []
    description_list = []
    price_list = []
    cost_list = []

    # @TODO: Loop over every row in the menu list object
    for row in csv_reader:
         # Item,Category,Description,Price,Cost
         # @TODO: Initialize menu data variables
        item =  row[0]
        category = row[1]
        description = [2]
        price = float(row[3])
        cost = float(row[4])
           
        item_list.append(item)
        category_list.append(category)
        description_list.append(description)
        price_list.append(price)
        cost_list.append(cost)

    menu_list.append(item_list)
    menu_list.append(category_list)
    menu_list.append(description_list)
    menu_list.append(price_list)
    menu_list.append(cost_list)

# @TODO: Set file paths for sales_data.csv
sales_filepath = Path('../PyRamen/sales_data.csv')

# @TODO: Initialize list objects to hold our sales data
sales_list = []
    
# @TODO: Read in the sales data into the sales list
with open(sales_filepath, 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter = ',')
    next(csv_reader, None) 

# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
    line_item_list =[]
    date_list =[]
    card_number_list =[]
    quantity_list = []
    sales_item_list = []

    
# Initialize a row counter variable


    # @TODO: Loop over every row in the sales list object
    for row in csv_reader:
        # Line_Item_ID,Date,Credit_Card_Number,Quantity,Menu_Item
        # @TODO: Initialize sales data variables
        line_item = row[0]
        date = row[1]
        card_number = row[2]
        quantity = int(row[3])
        sales_item = row[4]
              
        line_item_list.append(line_item)
        date_list.append(date)
        card_number_list.append(card_number)
        quantity_list.append(quantity)
        sales_item_list.append(sales_item)
    
    sales_list.append(line_item_list)
    sales_list.append(date_list)
    sales_list.append(card_number_list)
    sales_list.append(quantity_list)
    sales_list.append(sales_item_list)
    
    
report = {}

for sales_item in sales_item_list:
    
    # @TODO:
    # If the item value not in the report, add it as a new entry with initialized metrics
    if sales_item not in report:
        report[sales_item] = {
            '01-count' : 0,
            '02-revenue' : 0,
            '03-cogs': 0,
            '04-profit' : 0}
    
    # Naming convention allows the keys to be ordered in logical fashion, count, revenue, cost, profit

            

    # @TODO: For every row in our sales data, loop over the menu records to determine a match
for item, price, cost in zip(item_list, price_list, cost_list):
    for sales_item, quantity in zip(sales_item_list, quantity_list):
        
        # @TODO: If the item value in our sales data is equal to the any of the items in the menu, then begin tracking metrics for that item
        if item == sales_item:
            
             # @TODO: Cumulatively add up the metrics for each item key
            report[sales_item]["01-count"] += quantity
            report[sales_item]["02-revenue"] += price * quantity
            report[sales_item]["03-cogs"] += cost * quantity
            report[sales_item]["04-profit"] += (price - cost) * quantity

        # @TODO: Else, the sales item does not equal any fo the item in the menu data, therefore no match

    
# @TODO: Print out matching menu data    
print(report)

# @TODO: Calculate profit of each item in the menu data         

profit_list = []

with open(menu_filepath, 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter = ',')
    next(csv_reader, None)
    

    profit_item_list = []

    for row in csv_reader:
        profit_item = float(row[3]) - float(row[4])

        profit_item_list.append(profit_item)
        
    profit_list.append(profit_item_list)
    profit_dict = dict(zip(item_list,profit_item_list))
    print(profit_dict)
    
    # @TODO: Increment the row counter by 1


# @TODO: Print total number of records in sales data
row_count = sum((1 for line in open(sales_filepath)), -1)
print(row_count)


# @TODO: Write out report to a text file (won't appear on the command line output)
output = 'ramen_metrics.txt'

with open(output, 'w') as file:
    file.write(str(report))

```
